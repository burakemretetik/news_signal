```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import time
import os

def fetch_bist_stock_dataframes(stock_codes_string, period="max", interval="1d", auto_adjust=True, save_to_csv=True, output_folder="bist_data"):
    """
    Fetch historical data for BIST stocks and return as DataFrames.

    Parameters:
    stock_codes_string (str): Comma-separated string of stock codes
    period (str): Period of historical data to fetch (e.g., "max", "1y", "6mo")
    interval (str): Data interval (e.g., "1d", "1wk", "1mo")
    auto_adjust (bool): Adjust OHLC automatically
    save_to_csv (bool): Whether to save data to CSV files
    output_folder (str): Folder to save the CSV files (if save_to_csv is True)

    Returns:
    tuple: (dict of DataFrames with stock codes as keys, list of DataFrames, summary dict)
    """
    # Create output folder if saving to CSV and folder doesn't exist
    if save_to_csv and not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Split the string into a list of stock codes
    stock_codes = [code.strip() for code in stock_codes_string.split(',')]
    print(f"Found {len(stock_codes)} stock codes to process")

    # Prepare containers for results
    stock_dataframes = {}  # Dictionary with stock code as key and DataFrame as value
    all_dataframes = []    # List of all DataFrames

    # Prepare results tracking
    results = {
        "successful": [],
        "failed": []
    }

    # Get today's date for the filename if saving
    today = datetime.now().strftime("%Y-%m-%d")

    # Process each stock code
    for i, code in enumerate(stock_codes):
        try:
            # For BIST stocks, we need to append .IS to the ticker
            ticker_symbol = f"{code}.IS"
            print(f"Processing {i+1}/{len(stock_codes)}: {ticker_symbol}")

            # Create Ticker object
            ticker = yf.Ticker(ticker_symbol)

            # Fetch the historical data
            hist_data = ticker.history(
                period=period,
                interval=interval,
                auto_adjust=auto_adjust
            )

            if hist_data.empty:
                print(f"No data found for {ticker_symbol}")
                results["failed"].append(code)
                continue

            # Add stock code as a column for identification when in a list
            hist_data['StockCode'] = code

            # Store DataFrame in the dictionary and list
            stock_dataframes[code] = hist_data
            all_dataframes.append(hist_data)

            # Save to CSV if requested
            if save_to_csv:
                filename = f"{output_folder}/{code}_{today}.csv"
                hist_data.to_csv(filename)
                file_path = filename
            else:
                file_path = "Not saved"

            # Add to successful list
            results["successful"].append({
                "code": code,
                "rows": len(hist_data),
                "date_range": f"{hist_data.index.min().date()} to {hist_data.index.max().date()}" if not hist_data.empty else "N/A",
                "file": file_path
            })

            # Sleep briefly to avoid hitting rate limits
            time.sleep(0.5)

        except Exception as e:
            print(f"Error processing {code}: {str(e)}")
            results["failed"].append(code)

    # Print summary
    print("\nDownload Summary:")
    print(f"Successfully downloaded data for {len(results['successful'])} stocks")
    print(f"Failed to download data for {len(results['failed'])} stocks")

    # Create a summary CSV if saving files
    if save_to_csv:
        summary_df = pd.DataFrame(results["successful"])
        if not summary_df.empty:
            summary_df.to_csv(f"{output_folder}/download_summary_{today}.csv", index=False)

    return stock_dataframes, all_dataframes, results

# Example usage
if __name__ == "__main__":
    # Your comma-separated string of BIST stock codes
    bist = bist # Replace with your full list

    # Fetch the data
    stock_dfs, all_dfs_list, results = fetch_bist_stock_dataframes(
        bist,
        period="max",
        interval="1d",
        auto_adjust=True,
        save_to_csv=True  # Set to False if you don't want to save CSV files
    )

    # Example of accessing a single stock's DataFrame
    if "VAKBN" in stock_dfs:
        vakbn_df = stock_dfs["VAKBN"]
        print("\nVAKBN DataFrame sample:")
        print(vakbn_df.head())

    # Example of working with all stocks together
    if all_dfs_list:
        # Combine all DataFrames into one big DataFrame
        all_stocks_df = pd.concat(all_dfs_list)
        print("\nCombined DataFrame info:")
        print(f"Total rows: {len(all_stocks_df)}")
        print(f"Unique stocks: {all_stocks_df['StockCode'].nunique()}")

    # Print failures for manual review
    if results["failed"]:
        print("\nFailed downloads:")
        for code in results["failed"]:
            print(code)
```

```python
def analyze_stock_availability(stock_dfs):
    """
    Analyze the availability and historical depth of stock data.

    Parameters:
    stock_dfs (dict): Dictionary of stock DataFrames with stock codes as keys

    Returns:
    pandas.DataFrame: Summary of data availability for each stock
    """
    # Create a list to store availability data
    availability_data = []

    # Process each stock
    for code, df in stock_dfs.items():
        if df.empty:
            continue

        # Get the date range
        start_date = df.index.min().date()
        end_date = df.index.max().date()

        # Calculate years of data available
        years_available = (end_date - start_date).days / 365.25

        # Count total trading days
        trading_days = len(df)

        # Calculate data completeness (trading days vs. calendar days)
        calendar_days = (end_date - start_date).days
        completeness_pct = (trading_days / calendar_days * 100) if calendar_days > 0 else 0

        # Get average daily volume
        avg_volume = df['Volume'].mean() if 'Volume' in df.columns else None

        # Store the results
        availability_data.append({
            'StockCode': code,
            'Start_Date': start_date,
            'End_Date': end_date,
            'Years_Available': round(years_available, 2),
            'Trading_Days': trading_days,
            'Data_Completeness_Pct': round(completeness_pct, 2),
            'Avg_Daily_Volume': avg_volume
        })

    # Create a DataFrame from the results
    availability_df = pd.DataFrame(availability_data)

    # Sort by start date (oldest first)
    if not availability_df.empty:
        availability_df = availability_df.sort_values('Start_Date')

    return availability_df

def visualize_data_availability(availability_df, output_folder="bist_analysis"):
    """
    Create visualizations for stock data availability.

    Parameters:
    availability_df (pandas.DataFrame): DataFrame with availability information
    output_folder (str): Folder to save visualizations
    """
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set the style
    sns.set(style="whitegrid")

    # 1. Histogram of data age (years available)
    plt.figure(figsize=(10, 6))
    sns.histplot(availability_df['Years_Available'], bins=20, kde=True)
    plt.title('Distribution of Historical Data Availability')
    plt.xlabel('Years of Data Available')
    plt.ylabel('Number of Stocks')
    plt.savefig(f"{output_folder}/data_age_distribution.png")
    plt.close()

    # 2. Top 20 stocks with longest history
    plt.figure(figsize=(12, 8))
    top_historical = availability_df.nlargest(20, 'Years_Available')
    sns.barplot(x='Years_Available', y='StockCode', data=top_historical)
    plt.title('Top 20 Stocks with Longest Historical Data')
    plt.xlabel('Years of Data Available')
    plt.ylabel('Stock Code')
    plt.tight_layout()
    plt.savefig(f"{output_folder}/top_historical_stocks.png")
    plt.close()

    # 3. Timeline of data availability
    if len(availability_df) <= 50:  # Only for a reasonable number of stocks
        plt.figure(figsize=(12, 10))
        availability_df = availability_df.sort_values('Start_Date', ascending=False)

        for i, row in enumerate(availability_df.itertuples()):
            plt.plot([row.Start_Date, row.End_Date], [i, i], linewidth=2)
            plt.text(row.Start_Date, i, f" {row.StockCode}", verticalalignment='center')

        plt.yticks([])
        plt.title('Timeline of Data Availability by Stock')
        plt.xlabel('Date')
        plt.tight_layout()
        plt.savefig(f"{output_folder}/data_availability_timeline.png")
        plt.close()

    return

# Example usage with the previously fetched data
if __name__ == "__main__":
    # Assuming stock_dfs is already defined from the previous script
    # stock_dfs, all_dfs_list, results = fetch_bist_stock_dataframes(bist)

    # Analyze data availability
    availability_summary = analyze_stock_availability(stock_dfs)

    # Display the results
    print("\nStock Data Availability Summary:")
    print(f"Total stocks with data: {len(availability_summary)}")

    if not availability_summary.empty:
        # Show oldest stocks
        print("\nTop 10 stocks with oldest data:")
        print(availability_summary.head(10)[['StockCode', 'Start_Date', 'Years_Available', 'Trading_Days']])

        # Show newest stocks
        print("\nTop 10 most recent stocks (shortest history):")
        print(availability_summary.nsmallest(10, 'Years_Available')[['StockCode', 'Start_Date', 'Years_Available', 'Trading_Days']])

        # Summary statistics
        print("\nSummary statistics of data availability:")
        print(availability_summary[['Years_Available', 'Trading_Days', 'Data_Completeness_Pct']].describe())

        # Create visualizations
        visualize_data_availability(availability_summary)
        print("\nVisualizations saved to 'bist_analysis' folder")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
import os

def filter_stocks_since_date(stock_dfs, start_date='2015-01-01', output_folder="bist_filtered_data"):
    """
    Filter stocks to only include those with data available since the specified start date.

    Parameters:
    stock_dfs (dict): Dictionary of stock DataFrames with stock codes as keys
    start_date (str): Start date in 'YYYY-MM-DD' format
    output_folder (str): Folder to save filtered data

    Returns:
    tuple: (filtered dictionary of DataFrames, list of filtered DataFrames, summary DataFrame)
    """
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Convert start_date to datetime (without timezone)
    filter_date = pd.to_datetime(start_date)

    # Prepare containers for filtered data
    filtered_dfs = {}
    filtered_list = []
    summary_data = []

    # Process each stock
    for code, df in stock_dfs.items():
        if df.empty:
            continue

        # Check if data starts before or on the filter date
        # Handle timezone-aware DataFrames
        df_copy = df.copy()

        # Convert index to timezone-naive if it has timezone info
        if df_copy.index.tz is not None:
            df_copy.index = df_copy.index.tz_localize(None)

        earliest_date = df_copy.index.min()

        # Now compare with filter_date (both timezone-naive)
        if earliest_date <= filter_date:
            # Filter the DataFrame to only include data from the start date onwards
            filtered_df = df_copy[df_copy.index >= filter_date].copy()

            # Only include if there's meaningful data after filtering
            if not filtered_df.empty:
                # Add to filtered collections
                filtered_dfs[code] = filtered_df
                filtered_list.append(filtered_df)

                # Calculate data metrics for summary
                end_date = filtered_df.index.max().date()
                years_available = (end_date - filter_date.date()).days / 365.25
                trading_days = len(filtered_df)

                # Add to summary
                summary_data.append({
                    'StockCode': code,
                    'Start_Date': filter_date.date(),
                    'End_Date': end_date,
                    'Years_Available': round(years_available, 2),
                    'Trading_Days': trading_days,
                    'Data_Points': trading_days * (len(filtered_df.columns) - 1 if 'StockCode' in filtered_df.columns else len(filtered_df.columns))
                })

    # Create summary DataFrame
    summary_df = pd.DataFrame(summary_data)
    if not summary_df.empty:
        summary_df = summary_df.sort_values('StockCode')

        # Save summary to CSV
        today = datetime.now().strftime("%Y-%m-%d")
        summary_df.to_csv(f"{output_folder}/filtered_stocks_summary_{today}.csv", index=False)

    # Create a combined DataFrame if requested
    if filtered_list:
        combined_df = pd.concat(filtered_list)
        combined_df.to_csv(f"{output_folder}/all_filtered_stocks_{today}.csv")

    # Print summary
    print(f"\nFiltered Stocks Summary (since {start_date}):")
    print(f"Total stocks with data available: {len(filtered_dfs)} out of {len(stock_dfs)}")

    if not summary_df.empty:
        print(f"Average trading days per stock: {summary_df['Trading_Days'].mean():.1f}")
        print(f"Stocks with most trading days: {summary_df.nlargest(3, 'Trading_Days')['StockCode'].tolist()}")

    return filtered_dfs, filtered_list, summary_df

def save_filtered_dataframes(filtered_dfs, output_folder="bist_filtered_data"):
    """
    Save each filtered DataFrame to a CSV file.

    Parameters:
    filtered_dfs (dict): Dictionary of filtered stock DataFrames
    output_folder (str): Folder to save filtered data
    """
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get today's date for filenames
    today = datetime.now().strftime("%Y-%m-%d")

    # Save each DataFrame
    for code, df in filtered_dfs.items():
        filename = f"{output_folder}/{code}_since_2015_{today}.csv"
        df.to_csv(filename)

# Example usage
if __name__ == "__main__":
    # Assuming stock_dfs is already defined from the previous script
    # stock_dfs, all_dfs_list, results = fetch_bist_stock_dataframes(bist)

    # Filter stocks since January 1, 2015
    filtered_dfs, filtered_list, summary = filter_stocks_since_date(
        stock_dfs,
        start_date='2015-01-01'
    )

    # Save each filtered DataFrame to a separate CSV
    save_filtered_dataframes(filtered_dfs)

    # Show some sample data
    if filtered_dfs:
        sample_code = list(filtered_dfs.keys())[0]
        print(f"\nSample data for {sample_code}:")
        print(filtered_dfs[sample_code].head())
```

```python
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)

# Create a new DataFrame with stock codes as columns and Close prices as values
# First, get a list of all dates from all DataFrames
all_dates = set()
for df in filtered_dfs.values():
    all_dates.update(df.index.tolist())

# Sort dates
all_dates = sorted(all_dates)

# Create an empty DataFrame with dates as index
close_matrix = pd.DataFrame(index=all_dates)

# Fill in the Close prices for each stock
for code, df in filtered_dfs.items():
    close_matrix[code] = df['Close']

# Sort the DataFrame by date
close_matrix = close_matrix.sort_index()

# Display basic info about the resulting DataFrame
print(f"Shape of close_matrix: {close_matrix.shape}")
print(f"Number of dates: {len(close_matrix.index)}")
print(f"Number of stocks: {len(close_matrix.columns)}")
print(f"Date range: {close_matrix.index.min()} to {close_matrix.index.max()}")

# Display a sample of the data
print("\nSample of the close price matrix:")
print(close_matrix.head())

# Save the close price matrix to a CSV file
close_matrix.to_csv("bist_filtered_data/close_price_matrix.csv")
print("\nClose price matrix saved to 'bist_filtered_data/close_price_matrix.csv'")
```

```python
# First, let's check for NaN values in the close_matrix
nan_count_by_stock = close_matrix.isna().sum()
nan_count_by_date = close_matrix.isna().sum(axis=1)

# Calculate percentage of NaN values
total_cells = close_matrix.shape[0] * close_matrix.shape[1]
nan_cells = close_matrix.isna().sum().sum()
nan_percentage = (nan_cells / total_cells) * 100

print(f"Total missing values: {nan_cells} out of {total_cells} ({nan_percentage:.2f}%)")

# Find stocks with most missing values
stocks_with_most_nans = nan_count_by_stock.sort_values(ascending=False).head(10)
print("\nTop 10 stocks with most missing values:")
for stock, count in stocks_with_most_nans.items():
    print(f"{stock}: {count} missing values ({count/len(close_matrix.index)*100:.2f}%)")

# Find dates with most missing values
dates_with_most_nans = nan_count_by_date.sort_values(ascending=False).head(10)
print("\nTop 10 dates with most missing values:")
for date, count in dates_with_most_nans.items():
    print(f"{date}: {count} missing values ({count/len(close_matrix.columns)*100:.2f}%)")

# Check for gaps in the date index
# First, get the full range of business days from min to max date
full_date_range = pd.date_range(
    start=close_matrix.index.min(),
    end=close_matrix.index.max(),
    freq='B'  # Business days (Monday to Friday)
)

# Find missing dates (dates in full range but not in index)
missing_dates = set(full_date_range) - set(close_matrix.index)
extra_dates = set(close_matrix.index) - set(full_date_range)

print(f"\nMissing business days: {len(missing_dates)}")
if missing_dates:
    print("Sample of missing business days:")
    for date in sorted(list(missing_dates))[:10]:
        print(f"  {date.date()}")

print(f"\nDates in dataset that aren't standard business days: {len(extra_dates)}")
if extra_dates:
    print("Sample of non-standard business days in the data:")
    for date in sorted(list(extra_dates))[:10]:
        print(f"  {date.date()}")

# Analyze day-of-week distribution
day_of_week_counts = close_matrix.index.day_name().value_counts()
print("\nDay of week distribution:")
print(day_of_week_counts)

# Visualize the missing data pattern
plt.figure(figsize=(10, 6))
sns.heatmap(close_matrix.isna().sample(min(50, len(close_matrix))),
            cbar=False,
            yticklabels=False,
            cmap='viridis')
plt.title('Missing Value Pattern (Sample of 50 dates)')
plt.xlabel('Stocks')
plt.ylabel('Dates')
plt.savefig('bist_filtered_data/missing_data_pattern.png')

# Create a calendar heatmap to see missing data by date
def create_calendar_heatmap(nan_count_by_date):
    # Convert the Series to a DataFrame with year, month, day columns
    cal_data = pd.DataFrame({
        'nan_count': nan_count_by_date,
        'date': nan_count_by_date.index
    })
    cal_data['year'] = cal_data['date'].dt.year
    cal_data['month'] = cal_data['date'].dt.month
    cal_data['day'] = cal_data['date'].dt.day

    # Calculate missing percentage
    cal_data['missing_pct'] = cal_data['nan_count'] / len(close_matrix.columns) * 100

    # Create a pivot table for the heatmap
    years = sorted(cal_data['year'].unique())
    if len(years) > 3:
        # If too many years, limit to the most recent 3
        years = years[-3:]

    for year in years:
        year_data = cal_data[cal_data['year'] == year]
        pivot_data = year_data.pivot_table(
            index='day',
            columns='month',
            values='missing_pct',
            aggfunc='mean'
        )

        plt.figure(figsize=(12, 8))
        sns.heatmap(pivot_data, cmap='YlOrRd', vmin=0, vmax=100)
        plt.title(f'Percentage of Missing Values by Day - {year}')
        plt.xlabel('Month')
        plt.ylabel('Day')
        plt.savefig(f'bist_filtered_data/missing_data_calendar_{year}.png')

try:
    create_calendar_heatmap(nan_count_by_date)
    print("\nCalendar heatmaps saved to bist_filtered_data/ folder")
except Exception as e:
    print(f"Couldn't create calendar heatmap: {str(e)}")

# Check for continuous missing days
continuous_missing = []
current_streak = []
previous_date = None

for date, count in nan_count_by_date.sort_index().items():
    if count / len(close_matrix.columns) > 0.5:  # More than 50% stocks missing
        if previous_date is not None and (date - previous_date).days == 1:
            current_streak.append(date)
        else:
            if len(current_streak) >= 3:  # Only report streaks of 3+ days
                continuous_missing.append(current_streak)
            current_streak = [date]
    else:
        if len(current_streak) >= 3:
            continuous_missing.append(current_streak)
        current_streak = []

    previous_date = date

if len(current_streak) >= 3:
    continuous_missing.append(current_streak)

if continuous_missing:
    print("\nPeriods with continuous missing data (>50% stocks missing):")
    for streak in continuous_missing:
        print(f"  {streak[0].date()} to {streak[-1].date()} ({len(streak)} days)")
```
